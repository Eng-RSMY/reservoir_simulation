```python
import os
import pandas as pd
import matplotlib.pyplot as plt

total_resource = pd.DataFrame()
total_reserve = pd.DataFrame()
adsorbed_resource = pd.DataFrame()
adsorbed_reserve = pd.DataFrame()

for root_path, directories, files in os.walk('.'):
    if any(x in root_path for x in ['Ads_Langmuir_pressure','Ads_Langmuir_volume','Fracture_height','Fracture_permeability','Fracture_spacing','Fracture_width','Pore_Density','Pore_Permeability','Pore_Porosity']):
        if r'production.csv' in files:
            df = pd.read_csv(os.path.join(root_path,'production.csv'),header=None)
            col = root_path.split('\\')[-2]
            row = root_path.split('\\')[-1][-3:]
            total_resource.loc[row,col] = df.iloc[0,1]
            total_reserve.loc[row,col] = df.iloc[-1,3]
            adsorbed_resource.loc[row,col] = df.iloc[1,2]
            adsorbed_reserve.loc[row,col] = df.iloc[-1,4]
            
total_resource.to_csv('total_resource.csv')
total_reserve.to_csv('total_reserve.csv')
adsorbed_resource.to_csv('adsorbed_resource.csv')
adsorbed_reserve.to_csv('adsorbed_reserve.csv')
```
